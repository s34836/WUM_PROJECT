```python
import pandas as pd
import numpy as np
import warnings

# Suppress all runtime warnings (e.g., divide by zero, overflow)
warnings.filterwarnings("ignore", category=RuntimeWarning)
np.seterr(all='ignore')

cars_LinearRegression = pd.read_parquet("cars_ready_LinearRegression.parquet")
print("Price" in cars_LinearRegression.columns)
print(cars_LinearRegression.columns.tolist())


cars_DecisionTree = pd.read_parquet("cars_ready_DecisionTree.parquet")
print("Price" in cars_DecisionTree.columns)
print(cars_DecisionTree.columns.tolist())


cars_BART = pd.read_parquet("cars_ready_BART.parquet")
print("Price" in cars_BART.columns)
print(cars_BART.columns.tolist())
```

```python
# --- Filtracja najczęstszych cech wyposażenia ---
equipment_cols = [col for col in cars_LinearRegression.columns if col.startswith("Equipment_")]

top_50_equipment = (
    cars_LinearRegression[equipment_cols]
    .sum()
    .sort_values(ascending=False)
    .head(50)
    .index
    .tolist()
)

print("Top 50 equipment features selected:")
print(top_50_equipment)

# --- Zbiór wspólnych grup cech ---
embedding_cols = [col for col in cars_LinearRegression.columns if col.startswith("desc_emb_")]
binary_features = ["No_Accidents", "Service_Record", "Is_Imported", "First_Owner", "Professional_Seller"]
continuous_features = [
    "Mileage", "Log_Mileage", "Age", "Log_Age",
    "Mileage_per_Year", "Engine_Power", "Engine_Capacity", "Power_per_Liter"
]
# Cechy kategorialne już zakodowane one-hot (w cars_LinearRegression)
onehot_features = [
    col for col in cars_LinearRegression.columns
    if col.startswith(("Make_", "Make_Model_",  "Body_Type_", "Fuel_Type_", "Gearbox_", "Transmission_"))
]

# --- Target i meta ---
target_and_meta = ["Log_Price", "cv_fold", "split"]

# --- cars_ready_LinearRegression_small.parquet ---
cars_lr_small = cars_LinearRegression[
    continuous_features + binary_features + onehot_features + top_50_equipment + embedding_cols + target_and_meta
].copy()

# --- cars_ready_DecisionTree_small.parquet ---
categorical_features = ["Make", "Make_Model", "Body_Type", "Fuel_Type", "Gearbox", "Transmission"]

cars_tree_small = cars_DecisionTree[
    continuous_features + binary_features + categorical_features + top_50_equipment + embedding_cols + target_and_meta
].copy()

```

```python
print(cars_LinearRegression.shape)
print(cars_DecisionTree.shape)
print(cars_BART.shape)

print(cars_lr_small.shape)
print(cars_tree_small.shape)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# --- Krok 1: Identyfikacja kolumn embeddingów ---
desc_emb_cols = [col for col in cars_lr_small.columns if col.startswith("desc_emb_")]

# --- Krok 2: Dopasuj PCA na LinearRegression zbiorze ---
pca = PCA(n_components=50, random_state=42)
lr_emb_reduced = pca.fit_transform(cars_lr_small[desc_emb_cols])
tree_emb_reduced = pca.transform(cars_tree_small[desc_emb_cols])  # ten sam model

# --- Krok 3: Utwórz DataFrame z nazwami kolumn ---
pca_cols = [f"desc_pca_{i}" for i in range(50)]
lr_emb_df = pd.DataFrame(lr_emb_reduced, columns=pca_cols, index=cars_lr_small.index)
tree_emb_df = pd.DataFrame(tree_emb_reduced, columns=pca_cols, index=cars_tree_small.index)

# --- Krok 4: Usuń oryginalne embeddingi i dodaj PCA ---
cars_lr_small_pca = cars_lr_small.drop(columns=desc_emb_cols).join(lr_emb_df)
cars_tree_small_pca = cars_tree_small.drop(columns=desc_emb_cols).join(tree_emb_df)

# --- Krok 5: Potwierdź kształt danych po redukcji ---
print(f"cars_lr_small_pca shape: {cars_lr_small_pca.shape}")
print(f"cars_tree_small_pca shape: {cars_tree_small_pca.shape}")
```
